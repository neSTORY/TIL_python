```python
fp = open ("lang.html",mode="r")
fp
```

```python
# li 태그가 있는 문장 찾는 방법들

soup.select_one("ul > li")
soup.select_one("ul > li#py")
soup.select_one("li#py")
```

```python
soup.select_one("#lang > #py") # id 명이 lang인 태그에서 id명이 cp인 태그로 찾기
soup.select_one("ul#lang > li#py")
soup.select_one("#lang #py")
soup.select_one("li[id='py']")
soup.select_one("li:nth-of-type(2)") # li태그에 해당되는 데이터들 중에서 2번째
soup.select_one("li:nth-child(2)") # li 태그에 해당되는 데이터들 중에서 2번째 (브라우저에 따라 안될수도 있음)
```

```python
soup.select("li") # 리스트
soup.select("li")[1].string # 리스트에서 1번 인덱스
soup.find_all("li")[1].string
```

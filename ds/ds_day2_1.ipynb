```python
from bs4 import BeautifulSoup 
import re # 정규 표현식을 사용할 때 --- (※1)
html = """
<ul>
  <li><a href="hoge.html">hoge</li>
  <li><a href="https://example.com/fuga">fuga*</li>
  <li><a href="https://example.com/foo">foo*</li>
  <li><a href="http://example.com/aaa">aaa</li>
</ul>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
ss = soup.select("li")
for i in range(len(ss)):
    print(ss[i].string)
```

```python
[soup.select("li")[i].string for i in range(len(soup.select("li")))]
```

```python
tag_a = soup.select("a")
```

```python
[i.attrs["href"] for i in tag_a]
```

```python
res = soup.select('li > a')
addr = [ i.attrs['href'] for i in res]
addr
```

```python
# addr에서 https로 시작되는 주소출력1

for i in addr:
    if i[0:8] == "https://":
        print(i)
```

```python
# addr에서 https로 시작되는 주소출력2

import re
for i in range(len(addr)):
    res = re.match("https",addr[i])
    if res:
        print(addr[i])
```

```python
li = soup.find_all(href=re.compile("^https://"))
```

```python
li
for i in li:
    print(i.attrs["href"])
```

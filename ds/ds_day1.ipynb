```python
num1 = [12, 17, 19, 17, 23]
num2 = [26, 37, 26, 37, 91]
num3 = [10,10,10,10,10]
def mode(num):
    if len(num) != len(set(num)):
        cnt_mode= []
        cnt = 0
        for i in num:
            if num.count(i)>1:
                cnt_mode.append(i)
        
        if len(set(cnt_mode)) == 1:
            print(str(num)+" 의 최빈값은  "+str(cnt_mode[0]))
        else:
            cnt_mode = list(set(cnt_mode))
            print(str(num)+" 의 최빈값은 "+str(cnt_mode))
    else:
        print(str(num)+" 의 최빈값 존재하지 않습니다.")

mode(num1)
mode(num2)
mode(num3)
```

```python
url="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
```

```python
myarg={"stnId":"184"}
```

```python
#http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=184
```

```python
import urllib.parse as up
```

```python
params=up.urlencode(myarg)
addr=url+"?"+params
```

```python
addr
```

```python
res=urllib.request.urlopen(addr).read()
#클 -> 서버로 요청 -> 답변(200(정상), 40x(주소오류), 50x(서버오류)) =>
#서버는 클라이언트에게 메인페이지의 코드를 전송
```

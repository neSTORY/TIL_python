### 1번

3개의 각으로 삼각형의 예각, 직각, 둔각을 구별하는 프로그램을 만들어라.

```python
def check_tri(num):

    sum_Angle = 0

    for i in range(len(num)):
        sum_Angle += num[i]
    print(sum_Angle)

    if len(num) != 3:
        return print("삼각형이 아닙니다.(변의 문제)")

    elif sum_Angle != 180:
        return print("삼각형이 아닙니다.(각이 180도가 아님)")

    else:

        for i in range(3):
            if num[i] == 0:
                print("삼각형이 아닙니다.(각이 0도)")
                break
            elif num[i] > 90:
                print("둔각삼각형")
                break
            elif num[i] == 90:
                print("직각 삼각형")
                break

        else:
            print("예각삼각형")


num_list = [[60, 60, 60], [30, 60, 90], [20, 40, 120],
            [0, 90, 90], [60, 70, 80], [40, 40, 50, 50]]

for n in num_list:
    check_tri(n)
```

```python
def decriminate_triangle(list):
    sum = 0
    if len(list)!=3: #1 :각이 4개일때
        return "삼각형이 아니다"
    for i in range(len(list)): #2 : 각의 합이 180이 아닐때
        sum += list[i]
    if sum != 180:
        return "삼각형이 아니다"
    if 90 in list:
        return "직각삼각형"
    for i in list:
        if i>90:
            return "둔각삼각형"
    return "예각삼각형"

num_list = [[60, 60, 60], [30, 60, 90], [20, 40, 120],
            [0, 90, 90], [60, 70, 80], [40, 40, 50, 50]]

for n in num_list:
    check_tri(n)
```

### 3번
크롬드라이버 및 셀레니움을 사용하여 벅스,멜론, (지니)의 인기차트 데이터를 각각  50위까지 추출한 후, 모두 통합하여 데이터프레임으로 저장하고 엑셀로 저장하시오.

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd

# 멜론 접속
url = "https://www.melon.com/index.htm"
driver = webdriver.Chrome("chromedriver.exe")
html = driver.get(url)
time.sleep(2)

# 차트페이지 들어가기. 처음만 해봄
chart_button = driver.find_element_by_css_selector("#gnb_menu > ul:nth-child(1) > li.nth1 > a > span.menu_bg.menu01")
chart_button.click()
time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

title = soup.select("#frm > div > table > tbody > tr > td > div > div > div.ellipsis.rank01 > span > a")
name = soup.select("#frm > div > table > tbody > tr > td > div > div > div.ellipsis.rank02 > a")
rank = 1
song_data = []
for i in range(50):
    title_num = title[i].string
    name_num = name[i].string
    song_data.append(["Melon",rank, title_num, name_num])
    rank +=1

    
# 벅스
url = "https://music.bugs.co.kr/chart?wl_ref=M_left_02_01"
driver.get(url)
time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html,"html.parser")

title = soup.select(".title > a")
name = soup.select(".artist > a")
rank = 1
for i in range(50):
    title_num = title[i].string
    name_num = name[i].string
    song_data.append(["Bugs",rank, title_num, name_num])
    rank +=1
    
# 지니

url = "https://www.genie.co.kr/chart/top200"
driver.get(url)
time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html,"html.parser")

title = soup.select("a.title.ellipsis")
name = soup.select("a.artist.ellipsis")
rank = 1
for i in range(50):
    title_num = title[i].string.strip()
    name_num = name[i].string.strip()
    song_data.append(["Genie",rank, title_num, name_num])
    rank +=1

    
data = pd.DataFrame(song_data,columns = ["사이트", "순위", "제목", "가수"])

data.to_excel("music_chart_top50.xlsx",index = False)
```

```python
왜도(skew) : 어느정도로 비대칭인지?
- 음수인 경우 : 왼쪽이 긴 꼬리를 갖는 형태 # 중앙값이 오른쪽에 있음
- 양수인 경우 : 오른쪽이 긴 꼬리를 갖는 형태 # 중앙값이 왼쪽에 있음
- 0인 경우 : 평균과 중앙값이 같은 경우 # 완전 대칭 mean = median = mode(최빈수)
- -0.5 ~ 0.5 : 대체적으로 대칭적이다
- -1~-0.5, 0.5~1 : 어느정도 치우쳐 있다
- 그 이상의 차이면 많이 치우쳐 있다
```

```python
df.skew()
```

```python
첨도 : 확률분포의 뾰족한 정도 (kurt)
첨도가 3보다 작은 경우 확률분포가 높은 그래프를 그림
첨도가 3보다 높은 경우 완만한 그래프를 그림
첨도가 3보다 작다 => 특이값이 거의 없다
첨도가 3보다 크다 => 특이값이 크다
```

```python
df.kurt()
```

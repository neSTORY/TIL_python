```python
df = pd.DataFrame({'C1': np.random.randn(20),
 'C2': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a','b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b']})
df
```

```python
c1열 값에 대해 구간을 나누고자 함
최소값 ~ 최대값을 5개 구간을 균등하게 나눔
```

```python
df["C1"].min(), df["C1"].max()
```

```python
bins = np.linspace(df.C1.min(),df.C1.max(),5) # 균등구간
bins # 구간 설정
```

```python
df["C1bin"] = np.digitize(df["C1"], bins) # 구간에 해당하는 숫자로 바꿔줌(1 ~ 5 구간)
```

```python
1: -1.66878764 ~ -0.86648209 사이에 속함
2: -0.86648209 ~ -0.06417654 사이에 속함
    ...
5: 1.54043457 마지막 값
```

```python
df
```

```python
# groupby 객체 확인하기
df.groupby("C1bin").get_group(1) # 1번 그룹에 대한 정보
```

```python
# 각 구간별 데이터 건수 출력

df.groupby("C1bin")["C2"].value_counts()
df.groupby("C1bin").size()
df.groupby("C1bin").count()
```

```python
df["C1bin"] == 2
```

```python
# df의 C1bin열에 대해 get_dummies사용
pd.get_dummies(df["C1bin"], prefix="C1") # prefix 사용시 해당열 인덱스가 컬럼으로 감
```

```python
df
# c1열 값 c1열 값의 평균보다 크면 => "high"
# else => "low"
# high_low열에 저장
```

```python
mean = df["C1"].mean()
```

```python
high_low = []
mean = df["C1"].mean()
for c in df["C1"]:
    if c > mean:
        c = "high"
    else:
        c = "low"
    high_low.append(c)
df["high_low"] = high_low
df
```

```python
df.C1.mean()
```

```python
bins = np.linspace(df.C1.min(),df.C1.mean(),2)
bins
```

```python
high_low = []
bins = np.linspace(df.C1.min(),df.C1.mean(),2)
for i in np.digitize(df["C1"], bins):
    if i == 1:
        i = "low"
    else:
        i = "high"
    high_low.append(i)
df["high_low"] = high_low
```

```python
df["high_low"] = binarize(df["C1"].values.reshape(-1,1),threshold=df["C1"].mean())
df["high_low"].replace({1:"high",0:"low"},inplace=True)
df
```

```python
# np.where(조건, 참, 거짓)

df["high_low"] = np.where(df["C1"] > df["C1"].mean(), "high", "low")
df
```

```python
df.groupby("high_low").size()
df.groupby("high_low")["C1"].mean()
```

```python
df["high_low"] = np.where(df["C1"] > df["C1"].mean(), "high", "low")
df
```

```python
Q1 = np.percentile(df["C1"], 25)  # 1사분위수(25%)
Q3 = np.percentile(df["C1"], 75)  # 3사분위수(25%)

# IQR = Q3 - Q1
# outlier 경계선
# Q3 + IQR*1.5
# Q1 - IQR*1.5
```

```python
# np.where 안에 거짓 부분에 조건을 한번 더 넣어줘서 이중 if 문 처럼 만들 수 있다.

df["high_med_low"] = np.where(df["C1"]<Q1, "low",np.where(df["C1"]<Q3, "med", "high"))
df
```

```python
# 히트맵(열지도) : 2차원 데이터, 카테고리 형태
```

```python
data = titanic[["survived","pclass"]]
data
```

```python
sns.heatmap(data)
# 히트맵을 그릴땐 행렬구성으로 재지정을 해줘야함
```

```python
# 히트맵을 그리기 위해 데이터프레임 재구성(행렬 형태로)

data = titanic.pivot_table(index = "survived", columns = "pclass", aggfunc = "size")
data
```

```python
sns.heatmap(data, cmap = sns.light_palette("brown"), annot = True, fmt="d") # annot : 숫자 표시
# fmt : 포맷 지정 "d"-> 숫자

# 히트맵은 카테고리 형태를 보여줄 때 좋음
```

```python
# 실수, 종류가 섞여있는 데이터
# barplot, boxplot, pointplot, violinplot, stripplot, swarmplot
```

```python
tips
```

```python
# 요일별로 total_bill의 평균
# 요일별로 tip의 합계
```

```python
tips.groupby("day")["total_bill"].mean()
tips.groupby("day")["tip"].sum()
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, hue = "sex") # 평균을 보여줌. 선은 편차
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)

# 위에 점들은 outlier
# 사각형의 길이는 IQR
# 사각형의 꼭지점은 Q1, Q3
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips)
```

```python
# violinplot 에서 0밑에까지 표기된 이유는 뭘까?
# -> 값이 존재해서가 아니라 바이올린 형태를 그리기위해 표시된 것!
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True) # jitter=True : 겹쳐져있는걸 펼쳐서 보여줌 
```

```python
sns.swarmplot(x = "day", y = "total_bill", data = tips) # swarm은 많이 펼쳐서 보여줌
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex", data = tips)
# tips 데이터셋에서 요일별로 성별에 따른 total_bill의 평균을 시각화
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = tips)
```

```python
sns.violinplot(x = "day", y = "total_bill", hue = "sex", data = tips)
```

```python
# strptim 

# "문자열로된 날짜데이터" -> 날짜 객체 생성
# "2021-07-19 13:50"

dt.datetime.strptime("2021-07-19 13:50","%Y-%m-%d %H:%M")
```

```python
# timedelta, datedelta

dt1 = dt.datetime(2021,7,19,14)
dt2 = dt.datetime(2021,7,22,18)
```

```python
td = dt2-dt1
print(td)
```

```python
print(td.days)
print(td.seconds) # 총 시간 차이를 초로 환산 (4시간 * 3600)
print(td.total_seconds()) # 총 차이를 초로 환산 ((72시간(3일)+4시간) * 3600)
```

```python
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
```

```python
myarg = {"stnId" : "184"}
```

```python
# http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=184
```

```python
import urllib.parse as up
```

```python
params = up.urlencode(myarg)

addr = url + "?" + params
print(params, addr)
```

```python
res = urllib.request.urlopen(addr).read()

# 클라이언트 -> 서버요청 -> 답변(200(정상), 40x(주소오류), 50x(서버오류)) = >
# 서버는 클라이언트에게 메인
```

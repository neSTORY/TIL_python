```python
# 4.

def levenshtein(text_a, text_b):

    # 맨 처음칸이 0이므로 text[0]에 0 값을 대입하고 시작함
    text_a = list("0"+text_a)
    text_b = list("0"+text_b)

    # 레벤슈타인 거리를 구하기 위한 리스트 생성
    # 행렬화를 위해 리스트 안에 리스트를 넣은 구조
    leven = []
    for _ in range(len(text_a)):
        leven.append([])

    # 모든 칸에 0 채워넣기 -> 뒤에서 값 넣기 편하게
    for i in range(len(text_a)):
        for j in range(len(text_b)):
            leven[i].append(0)
    # 1행과 1열을 text 값에 맞게 초기화 하는 과정
    for i in range(len(text_b)):
        leven[0][i] = i

    for i in range(len(text_a)):
        leven[i][0] = i

#==============1행 1열 초기화 끝=========================

    # 위 3개 값과 비교하여 칸 채우기
    for i in range(1, len(text_a)):
        for j in range(1, len(text_b)):
            if text_b[j] != text_a[i]:
                leven[i][j] = min(leven[i-1][j-1], leven[i] # 3개 값 비교하여 가장 작은 값에서 1 더하기
                                  [j-1], leven[i-1][j]) + 1
            else:
                leven[i][j] = min(leven[i-1][j-1], leven[i] # 문자가 같으므로 가장 작은 값과 같은 값 갖기
                                  [j-1], leven[i-1][j])

    return leven[len(text_a)-1][len(text_b)-1]


test = ["한국", "코리아", "KOREA", "고려", "김대한"]

for i in range(len(test)):
    levenshtein("대한민국", test[i])
    short_distance = levenshtein("대한민국", test[0])
    short_string = test[0]
    if levenshtein("대한민국", test[0]) > levenshtein("대한민국", test[i]):
        short_distance = levenshtein("대한민국", test[i])
        short_string = test[i]

print("\"대한민국\"과 최소편집거리를 갖는 문자열은 %s이고 거리는 %s 입니다."
      % (short_string, short_distance))
```

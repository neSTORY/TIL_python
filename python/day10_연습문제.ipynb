```python
# 1.

num1 = [12, 17, 19, 17, 23]
num2 = [26, 37, 26, 37, 91]
num3 = [28, 30, 32, 34, 144]
def mode(num):
    if len(num) != len(set(num)):
        cnt_mode= []
        cnt = 0
        for i in num:
            if num.count(i)>1:
                cnt_mode.append(i)
        
        if len(set(cnt_mode)) == 1:
            print(str(num)+" 의 최빈값은  "+str(cnt_mode[0]))
        else:
            cnt_mode = list(set(cnt_mode))
            print(str(num)+" 의 최빈값은 "+str(cnt_mode))
    else:
        print(str(num)+" 의 최빈값 존재하지 않습니다.")

mode(num1)
mode(num2)
mode(num3)
```

```python
# 6.

def OneEditApart(a, b):
    if len(a) != len(b):
        if len(a) < len(b):  # 삽입 시
            if set(a) | set(b) == set(b):
                c = set(b)-set(a)
                c = list(c)
                b = b.replace(c[0], "")
                if list(a) == list(b):
                    print("True(삽입)")
                else:
                    print("False(삽입)")
            else:
                print("False(삽입)")
        else:  # 제거 시
            if set(a) & set(b) == set(b):
                print("True(제거)")
            else:
                print("Flase(제거)")
    else:  # 변환 시
        if len(set(a) & set(b)) == len(a)-1:
            print("True(변환)")
        else:
            print("False(변환)")


OneEditApart("cat", "dog")  # = false
OneEditApart("cat", "cats")  # = true
OneEditApart("cat", "cut")  # = true
OneEditApart("cat", "cast")  # = true
OneEditApart("cat", "at")  # = true
OneEditApart("cat", "acts")  # = false
```

```python
# 9.

import re

origin_num = """영일영-구구칠8-일구팔사
0일영.칠칠칠팔.이삼사
영 일 칠 삼 칠 오 팔 이 팔 이
영일일 34구구 4 오 9 이"""

origin_num = origin_num.split("\n")
origin_num
```

```python
def check_real_num(num):
    
    # 1. 기호 제거
    only_num = re.sub("[ -.]","",num)
    
    # 2. 한글 -> 숫자
    hangeul_to_num = {"영":"0","일":"1", "이":"2", "삼":"3", "사":"4", "오":"5", "육":"6", "칠":"7", "팔":"8", "구":"9"}
    
    _to_num = ""
    for i in num:
        if not i.isdigit():
            _to_num += hangeul_to_num.get(i,"") # get( ) 키값 만족하면 밸류 리턴, 키값이 없으면 "" 리턴
        else:
            _to_num += i
    print(_to_num)
    
    # 3. 010과 그 외로 분리 후 매치되는 객체가 있는지 확인
    
    if _to_num[0:3] == "010": # 010
        pat = re.match("010[\d]{4}[\d]{4}",_to_num)
        if pat: # pat에 매치되는 객체가 있는지 확인
            print("True")
        else:
            print("False")
    else:                     # 011 ~ 019
        pat2 = re.match("01[\d][\d]{3,4}[\d]{4}",_to_num)
        if pat2: # pat2에 매치되는 객체가 있는지 확인
            print("True")
        else:
            print("False")

for i in origin_num:
    check_real_num(i)
```

```python
# 10.

def kaprekar_number(num):

    if num == 1:
        print("1")
        return True

    else:
        double_num = str(num**2)
        print(double_num)

        for i in range(1, len(double_num)):
            if int(double_num[:i]) + int(double_num[i:]) == num:
                return True
        return False
```

```python
print(kaprekar_number(1))  # 1² =1, 1 = 1
print(kaprekar_number(45))
print(kaprekar_number(297))   #297의 제곱 = 88209 => 88+209 = 297
print(kaprekar_number(3213))
```
